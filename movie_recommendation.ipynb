```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data=pd.read_csv('movies.csv')
movie_data.head()
```

```python
selected_features=['genres','keywords','overview','tagline','cast','crew','director']
# print(selected_features)

for feature in selected_features:
    movie_data[feature]=movie_data[feature].fillna('')

combined_features=''
for feature in selected_features:
    combined_features+=movie_data[feature]+' '
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
print(similarity)
similarity.shape
```

```python
# Model steps combined
movie_name=input('Enter the movie you like:')
list_of_all_titles=movie_data['title'].tolist()
close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=close_match[0]
ind_of_movie=movie_data[movie_data['title']==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[ind_of_movie]))
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print('Movies similar to',close_match,' for you:\n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title=movie_data[movie_data['index']==index]['title'].values[0]
    print(i,end='')
    print('.',title)
    i+=1
    if i>30:
        break
```
